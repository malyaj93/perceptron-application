**3) Implement a perceptron to recognize the letter A by producing output +1 for images of the letter A and -1 for images of all other letters.**

```python

# TRAINING_SET = dataset["TRAINING_SET"]
# TEST_SET = dataset["TEST_SET"]
TRAINING_SET = dataset.TRAINING_SET
TEST_SET = dataset.TEST_SET

class Perceptron:
  trainingData=[]
  weights=[5]*36 #added extra one for bias. will consider it to be in the beginning
  alpha=0.01
  predictChar=''

  def __init__(self,predictChar):
    self.predictChar=predictChar

  def testPrint(self):
    print(self.target)
  
  def train(self,fonts):
    fontset= fonts[:]

    # # add bias 
    # if len(fontset[0][0])<36 : #checks if bias already added
    #   for font in fontset:
    #     font[0].insert(0,1)
    # # //// add bias
      

    loss=10000
    while loss > 0:
      lossTemp=0
      for font in fontset:
        fontCopy=font[0][:]
        fontCopy.insert(0,1)
        prediction = self.predictValue(fontCopy) # get predicted vale
        
        # decide target value
        target=0
        if font[1]==self.predictChar:
          target = 1
        else:
          target = -1

        # //// decide target value

        # calc Loss summation
        error = target - prediction
        lossTemp += error ** 2


        # Edit Weights
        
        alphaXerror = self.alpha * error
        for index, weight in enumerate(self.weights):
          self.weights[index] = weight + alphaXerror * fontCopy[index]

        # //// Edit Weights 
      loss=lossTemp

    print("Loss=========", loss)
 

  def predictValue(self,font):
    sumWX=0
    for index,bit in enumerate(font):
      sumWX+= bit*self.weights[index]

    prediction=0
    if sumWX < 0:
      prediction=-1
    else:
      prediction=1
    
    return prediction


  def testingSet(self, font):
    fontTest= font[:]
    fontTest.insert(0,1)
    prediction = self.predictValue(fontTest) # get predicted vale
    return prediction

    # # add bias 
    # if len(fontset[0][0])<36 : #checks if bias already added
    #   for font in fontset:
    #     font[0].insert(0,1)
    # # //// add bias
    # for font in fontset:
    #     prediction = self.predictValue(font[0]) # get predicted vale
    #     print(f'{font[1]}== {prediction}')

    

print("=============  TRAIN A  =====================")
perceptron_A = Perceptron('A')
perceptron_A.train(TRAINING_SET)
```
